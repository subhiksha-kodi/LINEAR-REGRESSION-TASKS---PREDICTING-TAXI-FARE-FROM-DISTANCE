LINEAR REGRESSION (PREDICTING TAXI FARE FROM DISTANCE)

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('regression_datasets_large.xlsx - Taxi_Fare.csv')

X=df[['Distance_km']]     #Machine learning models always expect X to be 2D. [] --> 1D, [[]] -->2D
Y=df['Fare_Rs']

model=LinearRegression()
model.fit(X,Y)

trip_distance=15
#Whenever you pass input X to model.predict(), even if it’s just a single sample, you must provide it as a 2D array (rows × features).
predicted_fare=model.predict([[trip_distance]])

print(f'Predicted fair for {trip_distance} km is Rs. {predicted_fare[0]:.2f}')
print(f'Base Fare (Fixed cost): Rs. {model.intercept_:.2f}')
print(f'Fare per km: Rs. {model.coef_[0]:.2f}')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X,Y,color='red',label='Actual Fares')
plt.plot(X,model.predict(X),color='blue',linewidth=2,label='Regression Line (Predicted Fares)')
plt.xlabel('Distance (km)')
plt.ylabel('Taxi Fare Prediction')
plt.legend()
plt.grid()
plt.savefig('plots/linear_regression_graph.png')
plt.show()
```
